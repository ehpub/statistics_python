## 정규 분포를 따르는 샘플 생성하기

- 20XX년 A시 성인 남성의 평균 키는 174cm이고 표준 편차는 7이라고 한다.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
mu = 174 #평균
std = 7 #표준편차
n = 100 #샘플 수
samples = norm.rvs(size=n,loc=mu, scale =std)
samples
```

```python
sd = pd.Series(samples)
sd.describe()
```

```python
min_val = int(np.min(samples))-1
max_val = int(np.max(samples))+1
bins = range(min_val,max_val+1)
plt.hist(samples,bins=bins)
plt.title(f'sample count = {n}')
plt.show()
```

- 샘플의 분포를 도식화

```python
def show_norm_samples_hist(mu = 0,std = 1,n = 1000):
  samples = norm.rvs(size=n,loc=mu, scale =std)
  min_val = int(np.min(samples))-1
  max_val = int(np.max(samples))+1
  bins = range(min_val,max_val+1)
  plt.hist(samples,bins=bins)
  plt.title(f'sample count = {n}')
  plt.show()
```

```python
import math
for i in range(2,9):
  n = int(math.pow(10,i))
  show_norm_samples_hist(mu=174, std=7, n=n)
```

- 샘플의 분포를 표준 점수로 변환 후 도식화

zscore = (sample - 평균)/표준 편차

```python
def show_norm_samples_z_hist(mu = 0,std = 1,n = 1000):
  samples = norm.rvs(size=n,loc=mu, scale =std)
  z_scores = (samples - mu) /std
  bins = np.arange(-6,7,0.1)
  plt.hist(z_scores,bins=bins)
  plt.title(f'sample count = {n}, zscore')
  plt.show()
```

```python
import math
for i in range(2,9):
  n = int(math.pow(10,i))
  show_norm_samples_z_hist(mu=174,std=7,n=n)
```

## 표준 정규 분포를 따르는 샘플 생성하기

평균이 0, 표준 편차가 1인 정규 분포를 표준 정규 분포라 한다.

norm.rvs 함수에 평균(loc)과 표준편차(scale)의 디폴트 값이 0과 1입니다. 따라서 인자를 전달하지 않으면 표준 정규 분포를 따르는 샘플을 생성합니다.

```python
samples = norm.rvs(size=100) #평균이 0이고 표준편차가 1일 정규 분포 샘플을 생성
samples
```

```python
sd = pd.Series(samples)
sd.describe()
```

```python
def show_stdnorm_samples_hist(n = 1000):
  samples = norm.rvs(size=n)
  bins = np.arange(-6,7,0.1)
  plt.hist( samples  ,bins=bins)
  plt.title(f'sample count = {n}')
  plt.show()
```

```python
for i in range(2,9):
  n = int(math.pow(10,i))
  show_stdnorm_samples_hist(n=n)
```

## 확률 밀도 함수

확률 변수의 분포를 나타내기 위한 함수

확률 변수 X가 구간 [a,b]에 포함될 확률을 구하는 함수입니다.

```python
help(norm.pdf)
```

- 정규 분포를 따르는 샘플과 확률 밀도 함수를 같이 그리기

```python
def show_norm_samples_pdf(mu = 0,std = 1,n = 1000):
  samples = norm.rvs(size=n,loc=mu, scale =std)
  min_val = int(np.min(samples))-1
  max_val = int(np.max(samples))+1
  bins = range(min_val,max_val+1)
  y = norm.pdf(bins,mu,std)*n
  plt.hist(samples,bins=bins,label='histogram')
  plt.plot(bins,y,label='pdf')
  plt.legend()
  plt.title(f'sample count = {n}')
  plt.show()
```

```python
import math
for i in range(2,9):
  n = int(math.pow(10,i))
  show_norm_samples_pdf(mu=174,std=7,n=n)
```

## 누적 밀도 함수

확률 변수가 특정 값보다 작거나 같을 확률을 구하는 함수



```python
help(norm.cdf)
```

- 평균 키가 174cm이고 표준 편차가 7일 때 170cm 이하일 확률을 구하시오.

```python
 norm.cdf(loc=174,scale=7,x=170)
```

- 정규 분포를 따르는 샘플과 누적 확률 밀도 함수를 같이 그리기

```python
def show_norm_samples_cdf(mu = 0,std = 1,n = 1000, value=mu):
  samples = norm.rvs(size=n,loc=mu, scale =std)
  min_val = int(np.min(samples))-1
  max_val = int(np.max(samples))+1
  bins = range(min_val,max_val+1)
  py = norm.pdf(bins,mu,std)*n
  cy = norm.cdf(loc=mu,scale=std,x=value)
  plt.hist(samples,bins=bins,label='histogram')
  plt.plot(bins,py,label='pdf')

  bins2 =  range(min_val,int(value)+1)
  py2 = norm.pdf(bins2,mu,std)*n
  plt.fill_between(bins2,py2,alpha=0.7,color='green',label=f'less than={cy:.2f}')

  plt.axvline(value,color='r')
  plt.legend()
  plt.title(f'sample count = {n}')
  plt.show()
```

```python
import math
for i in range(2,9):
  n = int(math.pow(10,i))
  show_norm_samples_cdf(mu=174,std=7,n=n,value=170)
```

##  생존 확률

생존 확률 = 1 - cdf

```python
help(norm.cdf)
```

평균 키가 174cm이고 표준 편차가 7일 때 170cm 이하일 확률과 170cm 초과일 확률을 구하시오.

```python
a =  norm.cdf(loc=174,scale=7,x=170)
b = 1 -a
print(f"170 이하일 확률:{a:.2f}")
print(f"170 초과일 확률:{b:.2f}")
```

```python
a =  norm.cdf(loc=174,scale=7,x=170)
b = norm.sf(loc=174,scale=7,x=170)
print(f"170 이하일 확률:{a:.2f}")
print(f"170 초과일 확률:{b:.2f}")
```
