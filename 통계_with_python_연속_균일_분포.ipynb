- 샘플 생성하기

rvs(loc=0, scale=1, size=1, random_state=None)

```python
sample = uniform.rvs(size=10)
print(sample)
```

```python
for n in [100,1000,10000,100000]:
  sample = uniform.rvs(size=n)
  plt.hist(sample,bins=np.linspace(0,1,11),color='c',edgecolor='k')
  plt.title(f'size={n}')
  plt.show()
```

```python
sample = uniform.rvs(scale=5,size=10)
print(sample)
```

```python
for s in [1,2,5]:
  sample = uniform.rvs(scale=s,size=1000)
  plt.hist(sample,bins=np.linspace(0,s,11),color='c',edgecolor='k')
  plt.title(f'scale={s}')
  plt.show()
```

```python
sample = uniform.rvs(loc=5,scale=5,size=10)
print(sample)
```

```python
for m in [1,2,5]:
  sample = uniform.rvs(loc=m,scale=5,size=1000)
  plt.hist(sample,bins=np.linspace(m,m+5,11),color='c',edgecolor='k')
  plt.title(f'loc={m}, scale=5')
  plt.show()
```

- 확률 밀도 함수, 누적 분포 함수

pdf(x, loc=0, scale=1)

cdf(x, loc=0, scale=1)

```python
x = np.linspace(0,1,21)
y = uniform.pdf(x=x)
plt.plot(x,y,color='r')
plt.fill_between(x,y)
plt.title('uniform.pdf')
plt.show()
```

```python
x = np.linspace(0,1,21)
y = uniform.cdf(x=x)
plt.plot(x,y,color='r')
plt.fill_between(x,y)
plt.title('uniform.cdf')
plt.show()
```

A마을에 마을 버스는 1시간 주기로 다닌다고 한다.
버스 시간표를 모르는 사람이 A마을 버스 정류장에 갔을 때 대기시간이 7분 이내일 확률을 구하시오.

```python
y = uniform.cdf(scale=60,x=7)
print(y)
y = uniform.cdf(x=7/60)
print(y)
```

대기시간이 5분에서 20분 이내일 확률을 구하시오.

```python
ly = uniform.cdf(scale=60,x=5)
hy = uniform.cdf(scale=60,x=20)
print("대기시간이 5분에서 20분 이내일 확률은 ",hy-ly)
```

대기시간이 50분 이내 및 초과할 확률을 구하시오.

```python
y1 = uniform.cdf(scale=60,x=50)
y2 = uniform.sf(scale=60,x=50)
print("대기시간이 50분 이내일 확률은 ",y1)
print("대기시간이 50분 초과할 확률은 ",y2) #y2 = 1-y1
```

대기시간이 m분 이내일 확률이 73%라고 한다. m은 얼마인지 구하시오.

```python
m = uniform.ppf(scale=60,q=0.73)
print(f"대기 시간이 {m}분 이내일 확률은 73%")
```
